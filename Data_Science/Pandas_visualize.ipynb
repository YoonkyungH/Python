```python
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv('data/broadcast.csv', index_col=0)
df
```

```python
df.plot()
```

pandas dataframe에는 그래프 그리는 기능 내장.  
원래 plot에는 parameter로 kind를 넣어야함. (선 그래프를 그릴 때) kind='line'이 default 값이므로 이번엔 생략 가능. **df.plot(kind='line')**

```python
df.plot(y='KBS')
```

```python
df.plot(y=['KBS', 'JTBC'])
```

```python
df[['KBS', 'JTBC']]
```

```python
df[['KBS', 'JTBC']].plot()
```

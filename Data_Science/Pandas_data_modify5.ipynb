이렇듯 rename함수는 기존의 데이터를 바꾸지 않고 새로운 dataframe을 만들어버림.  
inplace=True를 쓰면 해결 가능

```python
liverpool_df.rename(columns={'position':'Position'}, inplace=True)
```
